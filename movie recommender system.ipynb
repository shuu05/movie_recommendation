# User based collaborative filtering

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
data = pd.merge(movies,ratings)
data.head()
```

```python
print(data['title'])
```

```python
data.drop('timestamp',axis=1)
```

```python
new_data = data.pivot_table(index=['userId'],columns=['title'],values='rating')
new_data.head(20)
```

```python
new_data.columns
```

```python
mv = new_data['xXx (2002)'] 
mv.head(10)
```

```python
similar_mv = new_data.corrwith(mv)
similar_mv = similar_mv.dropna()
df = pd.DataFrame(similar_mv)
df.head(10)
```

```python
similar_mv.sort_values(ascending=False)
```

# Item based collaorative filtering

```python
mv_stats = data.groupby('title').agg({'rating':[np.size,np.mean]})
```

```python
mv_stats.head()
```

```python
popular_mv = mv_stats['rating']['size'] >= 100
mv_stats[popular_mv].sort_values([('rating','mean')], ascending=False)[:15]
```

```python
df = mv_stats[popular_mv].join(pd.DataFrame(similar_mv, columns = ['similarity']))
```

```python
df.head(10)
```

```python
df.sort_values(['similarity'], ascending=False)[:15]
```
